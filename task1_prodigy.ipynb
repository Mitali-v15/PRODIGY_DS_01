```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset (Online population dataset from World Bank)
url = "https://datahub.io/core/population/r/population.csv"
df = pd.read_csv(url, encoding='utf-8')
```

```python
countries = ['United States', 'China', 'India', 'Brazil']
```

```python
for country in countries:
    country_data = df[df['Country Name'] == country]
```

```python
years = country_data['Year']
population = country_data['Value']
```

```python
plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {country} Over the Years')
plt.grid(True)
plt.show()
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
